```python
print("Rating_points:",mydata["Rating"].unique())
```

```python
my_counts =mydata["Rating"].value_counts()

print("total number of ratings:",mydata["Rating"].value_counts())
```

```python
sns.countplot(my_counts)
```

```python
plt.figure(1, figsize=(10,10))
my_counts.plot.pie(autopct="%1.1f%%")
```

```python
plt.scatter(mydata.Year, mydata.Rating, alpha = 0.28, label = "Movie", color = "blue")
plt.xlabel("Years")
plt.ylabel("Ratings")
plt.legend(loc = "lower right")
plt.show()
```

```python
#histogram plot about number of published movies according to year


mydata.Year.plot(kind = "hist", bins = 40, figsize = (9,6))
plt.xlabel("Years")
plt.ylabel("Number of Movies")
plt.show()
```
